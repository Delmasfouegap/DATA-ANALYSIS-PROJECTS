## Studiamo le relazioni tra le variabili

#### -*Qual è la distribuzione delle ore lavorate?*

```python
df["Work_Hours_Per_Week"].value_counts().sort_index(ascending=True).plot(kind="bar", title="distribution of working hours")
plt.show()
```

### Distribuzione dell'orario di lavoro
- La maggior parte dei dipendenti lavora tra le **32 e le 60 ore settimanali**.
- La distribuzione è relativamente uniforme, con una leggera tendenza al ribasso del numero di dipendenti all'aumentare delle ore.
- Ciò suggerisce che il carico di lavoro è generalmente omogeneo, ma alcuni dipendenti superano significativamente la norma (fino a 60 ore).
- Attenzione: queste variazioni potrebbero influenzare la soddisfazione o le prestazioni.

#### -*Qual è la correlazione tra ore lavorate e punteggio di prestazione?*

```python
corr = df["Work_Hours_Per_Week"].corr(df["Employee_Satisfaction_Score"], method="spearman")
print("la corelazione tra questi due variabile è:", corr)
sns.regplot(x="Work_Hours_Per_Week", y="Employee_Satisfaction_Score", data=df)
plt.title("Relationship between working hours and satisfaction")
plt.show()
```

### Relazione tra ore di lavoro e soddisfazione
- I punti sono distribuiti uniformemente verticalmente per ogni numero di ore → tutti i livelli di soddisfazione (1–5) coesistono.
- Nessuna tendenza visibile (né al rialzo né al ribasso): correlazione ≈ 0.
- Conclusione: le ore lavorate a settimana **non spiegano** il punteggio di soddisfazione in questo set di dati.

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df["Department"], df["Employee_Satisfaction_Score"])
chi2, p, dof, expected = chi2_contingency(contingency)
print("Chi²:", chi2, "p-value:", p)
```

### Relazione tra reparto e soddisfazione
- Test del chi-quadrato: Chi-quadrato = 3204,30, valore p = 0,475.
- Poiché il valore p è significativamente superiore a 0,05, non rifiutiamo l'ipotesi nulla.
- Conclusione: la soddisfazione dei dipendenti è **indipendente dal reparto** nella nostra azienda.

#### -*Ci sono reparti in cui soddisfazione e performance non sono in linea?*

```python
tab = pd.crosstab([df["Performance_Score"], df["Employee_Satisfaction_Score"]], df["Department"])
chi2, p, dof, expected = chi2_contingency(tab)
print("Chi²:", chi2, "p-value:", p)
```

### Relazione tra soddisfazione, performance e reparto
- Test del chi-quadrato: Chi-quadrato = 15.950,32, valore p = 0,675
- Poiché il valore p è > 0,05, l'ipotesi nulla non viene rifiutata.
- Conclusione: Non vi è **alcuna differenza significativa** tra soddisfazione e performance nei reparti.

```python
sns.scatterplot(x="Performance_Score", y="Employee_Satisfaction_Score", hue="Department", data=df)
plt.title("Performance vs. Satisfaction by Department")
plt.legend(labelcolor='grey', prop={'size': 10})
plt.show()
```

### Relazione tra performance e soddisfazione per reparto

- I dipendenti sono distribuiti sull'intera scala di soddisfazione (da 1 a 5), ​​indipendentemente dal punteggio di performance.
- La combinazione di colori per reparto mostra che la distribuzione è abbastanza uniforme: nessun reparto si distingue per una differenza marcata (ad esempio, elevata soddisfazione ma basse performance).
- I risultati confermano i test statistici (chi-quadrato e correlazione): **non esiste una relazione significativa tra performance e soddisfazione, né alcuna differenza particolare per reparto**.

#### -*le ore lavorate + la soddisfazione spiegano il punteggio di prestazione?*

```python
sns.lmplot(x="Work_Hours_Per_Week", 
           y="Employee_Satisfaction_Score", 
           col="Performance_Score", 
            data=df)
plt.Figure(figsize=(12,8))
plt.show()
```

### Le ore lavorate e la soddisfazione spiegano la performance?

- La visualizzazione sfaccettata (per punteggio di performance) mostra che:
- Le ore lavorate variano sempre tra 30 e 60 ore, indipendentemente dal livello di performance.
- La soddisfazione copre l'intera scala (da 1 a 5) per tutti i livelli di performance.
- La distribuzione dei punti è omogenea e non mostra **una tendenza chiara**.
- Conclusione: **né le ore lavorate né la soddisfazione sembrano spiegare direttamente il punteggio di performance**.
